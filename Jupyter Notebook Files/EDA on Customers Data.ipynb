```python
count = {}
for i in df['Profession']:
    if i in count:
        count[i]+=1
    else:
        count[i] = 1
```

```python
categories = count.keys()
values = count.values()
```

```python
plt.pie(values,labels= categories, autopct = "%1.1f%%")
plt.show()
```

```python
Gender_Count = {}
for i in df['Gender']:
    if i in Gender_Count:
        Gender_Count[i] += 1
    else:
        Gender_Count[i] = 1
```

```python
Gender_Count
```

```python
Gender_Label = Gender_Count.keys()
```

```python
Gender_Values = Gender_Count.values()
```

```python
plt.pie(Gender_Values,labels = Gender_Label, autopct = '%1.1f%%')
plt.show()
```
