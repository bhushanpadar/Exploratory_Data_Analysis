# Encoding Categorical Features

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
Label_Enc = LabelEncoder()
```

```python
df['Road_Class'] = Label_Enc.fit_transform(df['Road_Class'])
```

```python
for col in ['Lighting Conditions','Weather Conditions','Vehicle_Type','Casualty_Class','Casualty_Severity','Sex_of_Casualty']:
    df[col] = Label_Enc.fit_transform(df[col])
```

```python
df.head()
```

```python
df['Road Surface'] = Label_Enc.fit_transform(df['Road Surface'])
```
